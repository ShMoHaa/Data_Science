```python
# IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import linkage 
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
```

```python
#  IMPORTING DATASET
airlines=pd.read_csv("EastWestAirlines.csv")
airlines
```

```python
airlines.info()
```

```python
# STANDERDIZING 
scaler= StandardScaler()
standardize_data= pd.DataFrame(scaler.fit_transform(airlines.iloc[:,1:]))
standardize_data
```

```python
# NORMALIZING
normalized_data= pd.DataFrame(normalize(airlines.iloc[:,1:]))
normalized_data
```

# HIERARICHAL CLUSTERING

```python
z = linkage(normalized_data, method="complete",metric="euclidean")
```

```python
plt.figure(figsize=(20, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Features')
plt.ylabel('Airline')
sch.dendrogram(z,leaf_rotation=0., leaf_font_size=8)
plt.show()
```

```python
# CREATE CLUSTERS 
HC= AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```python
heir_df=airlines.copy()
heir_df['Cluster_ID']=HC.fit_predict(normalized_data)
heir_df.drop('ID#',axis=1,inplace=True)
```

```python
heir_df.groupby('Cluster_ID').mean()
```

```python
heir_df.Cluster_ID.value_counts()
```

# K-MEANS

```python
# CALCULATING WCSS
wcss= []
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=4)
    kmeans.fit(normalized_data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,10),wcss,"o-r")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,random_state=44)
kmeans.fit(normalized_data)
```

```python
km_df=airlines.copy()
km_df.drop('ID#',axis=1,inplace=True)
```

```python
km_df['Cluster_ID']= kmeans.labels_
km_df
```

```python
km_df.Cluster_ID.value_counts()
```

```python
km_df.groupby('Cluster_ID').mean()
```

# DBSCAN

```python
dbscan = DBSCAN(eps=0.08,min_samples=5)
dbscan.fit(normalized_data)
```

```python
clusters=pd.DataFrame(dbscan.labels_,columns=['Cluster_ID'])
clusters.value_counts()
```

```python
#VISUALISING CLUSTERS
colors=plt.get_cmap('viridis')
plt.figure(figsize=(15, 10))  
plt.scatter(km_df['Cluster_ID'],km_df['Balance'], c=kmeans.labels_,cmap=colors,edgecolors='BLUE') 
plt.xlabel("cluster_ID")
plt.ylabel("Balance")
```
