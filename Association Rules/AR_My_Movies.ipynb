```python
# IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# LOADING DATASET
data=pd.read_csv("my_movies.csv")
data
```

```python
data.info()
```

```python
final_data=data.iloc[:,5:]
final_data
#Droppping irrelevant  columns
```

# APRIORI ALGORITHM

## 1. Association rule with 10% Support and  90% confidence 

```python
frequent_itemsets1=apriori(final_data,min_support=0.1,use_colnames=True)
frequent_itemsets1
```

```python
rules1=association_rules(frequent_itemsets1,min_threshold=0.9,metric='lift')
rules1
```

```python
rules1.sort_values('lift',ascending = False)
```

```python
rules1[rules1.lift>1]
```

```python
# VISUALIZATION
plt.scatter(rules1['support'],rules1['confidence'],color='cyan')
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

## 2. Association with 20%  support and 60% confidence

```python
frequent_itemsets2=apriori(final_data,min_support=0.2,use_colnames=True)
frequent_itemsets2
```

```python
rules2=association_rules(frequent_itemsets2,min_threshold=0.6,metric='lift')
rules2
```
